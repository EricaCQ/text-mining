```python
import requests, xlrd, xlwt
```

```python
#Set the API´ keys resquest including the language variable. The keys are individual and personal, Ask for your keys at 
#https://forms.gle/wFjmy53sKnNmA4iX8
```

```python
LANG = 'pt'#Set Language
APIKEY = 'YOUR KEY for polarity test'
APIURL = 'https://sentic.net/api/' + LANG + '/' + APIKEY + '.py?text='
FILENAME = 'YOUR FILE' #Must be xls, since the xlrd method doesn´t aloow csv or txt.
```

```python
#SDK
wb = xlrd.open_workbook(FILENAME)
sheet = wb.sheet_by_index(0)
new_wb = xlwt.Workbook()
new_sheet = new_wb.add_sheet('labeled')

count = 0
for row in range(sheet.nrows):
    text = sheet.cell_value(row, 0)
    for c in [';', '&', '#', '{', '}']: text = text.replace(c, ':')
    label = str(requests.get(APIURL + text).content)[2:-3]
    new_sheet.write(count, 0, text)
    new_sheet.write(count, 1, label)
    count += 1
    print(text + ': ' + label)
    
new_wb.save(FILENAME + '_labeled.xls')
```
